```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
#bicep-high1 90-270
#bicep-down 100-235
#bicep-low1 110-end
#bicep-medium 90-end
#elbow-rot1 105-270
#shoulder1 90-last
#shoulder-down 110-end
#wrist 100-end
df1 = pd.read_csv('elbow_Rot1.csv')
data_np = df1.values
```

```python
scaler = MinMaxScaler()
linear_x=df1['field.linear.x'].values
linear_x = np.reshape(linear_x, (-1,1))
scaler.fit(linear_x)
ln_x_tr = scaler.transform(linear_x)
```

```python
scaler = MinMaxScaler()
linear_y=df1['field.linear.y'].values
linear_y = np.reshape(linear_y, (-1,1))
scaler.fit(linear_y)
ln_y_tr = scaler.transform(linear_y)
```

```python
scaler = MinMaxScaler()
linear_z=df1['field.linear.z'].values
linear_z = np.reshape(linear_z, (-1,1))
scaler.fit(linear_z)
ln_z_tr = scaler.transform(linear_z)
```

```python
scaler = MinMaxScaler()
angular_x=df1['field.angular.x'].values
angular_x = np.reshape(angular_x, (-1,1))
scaler.fit(angular_x)
an_x_tr = scaler.transform(angular_x)
```

```python
plt.plot(ln_x_tr, label='linear-x')
plt.plot(ln_y_tr, label='linear-y')
plt.plot(ln_z_tr, label='linear-z')
plt.plot(an_x_tr, label='angular-x')
plt.legend()
plt.show()
```
